## Basis Fit

```python
def FermiKernel(w, t, beta):
    x = beta*w/2
    y = 2*t/beta-1
    if x > 100:
        return np.exp(-x*(y+1.))
    if x < -100:
        return np.exp(x*(1.0-y))
    return np.exp(-x*y)/(2*np.cosh(x))
```

```python
Nw = 1000
w = np.linspace(-20, 20, Nw)
Nt = len(TauGrid)
kMatrix = np.zeros([Nw, Nt])

for i in range(len(w)):
    kMatrix[i, :] = FermiKernel(w[i], TauGrid, Para.Beta)

u, s, v = linalg.svd(kMatrix)
v_inv = linalg.inv(v)
```
